# create loop for each item

```python
def getelements():
    try:
        # basic page till iteration
        URL = "https://www.njuskalo.hr/prodaja-stanova?geo%5BlocationIds%5D=1576%2C1577%2C1578%2C1579"
        headers = {
'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.142 Safari/537.36'
 }
        page = requests.get(URL,headers=headers)
        soup = BeautifulSoup(page.content, "html.parser")
        wrap_main_el = soup.find_all("div", class_="wrap-main")
        wrap_content = wrap_main_el[0].find_all("div", class_="wrap-content ContentLayout ContentLayout--beta ContentLayout--browseFiltered cf")
        content_primary = wrap_content[0].find_all("div", class_="content-primary")
        main =  content_primary[0].find_all("main")
        form = main[0].find_all("form")
        wrap_content_primary = form[0].find_all("div", class_="wrap-content-primary cf")
        content_main = wrap_content_primary[0].find_all("div", class_="content-main")
        block_standard = content_main[0].find_all("div", class_="block-standard block-standard--epsilon")
        entity_list = block_standard[0].find_all("div", class_="EntityList--Standard")
        entity_list = entity_list[1:]
        vauvau_list = entity_list[0]
        vauvau_list = vauvau_list.find_all("ul", class_="EntityList-items")
        standard_list = entity_list[1]
        standard_list = standard_list.find_all("ul", class_="EntityList-items")
    except Exception as e:
        traceback.print_tb(e.__traceback__)
    
    df = pd.DataFrame(columns={'title', 'description','area_m2','location','price_eur','price_hrk'})
    df = df[['title', 'description','area_m2','location','price_eur','price_hrk']]
    #iteration
    try:        
        #vauvau list
        vauvau_list = vauvau_list[0].find_all("li", class_="EntityList-item--VauVau")
        for i in range(len(vauvau_list)):
            item = vauvau_list[i]
            article = item.find_all("article")[0]
            title = article.find_all("h3", class_="entity-title")[0].select("a")[0].text
            description = article.find_all("div", class_="entity-description")[0]
            description = list(description.select_one('.entity-description-main').stripped_strings)

            desc_1 = description[0]
            desc_2 = description[1]
            desc_3 = description[3]

            area_m2 = float(desc_2.split(':')[1].strip()[:-2].strip())
            price = article.find_all("div", class_="entity-prices")[0].select("strong")[0].text.strip()
            price_hrk = float((price.split('/')[1].split('\xa0')[0].strip().replace('.','')).replace(',','.'))
            price_eur = float(price.split('/')[0].split('\xa0')[0].strip().replace('.','').replace(',','.'))

            df.loc[len(df.index)] = [title, desc_1, area_m2, desc_3, price_eur, price_hrk] 
        #standard list
        standard_list = standard_list[0].find_all("li", class_="EntityList-item--Regular")
        for i in range(len(standard_list)):
            item_standard = standard_list[i]
            article_standard= item_standard.find_all("article")[0]
            title_standard = article_standard.find_all("h3", class_="entity-title")[0].select("a")[0].text
            description_standard = article_standard.find_all("div", class_="entity-description")[0]
            description_standard = list(description_standard.select_one('.entity-description-main').stripped_strings)
            desc_1_standard = description_standard[0]
            desc_2_standard = description_standard[1]
            desc_3_standard = description_standard[3]
            area_m2_standard = float(desc_2_standard.split(':')[1].strip()[:-2].strip())
            price_standard = article_standard.find_all("div", class_="entity-prices")[0].select("strong")[0].text.strip()
            price_hrk_standard = float((price_standard.split('/')[1].split('\xa0')[0].strip().replace('.','')).replace(',','.'))
            price_eur_standard = float(price_standard.split('/')[0].split('\xa0')[0].strip().replace('.','').replace(',','.'))

            df.loc[len(df.index)] = [title_standard, desc_1_standard, area_m2_standard, desc_3_standard, price_eur_standard, price_hrk_standard] 
    except Exception as e:
        traceback.print_tb(e.__traceback__)
    return df
```
